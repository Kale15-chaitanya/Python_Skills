```python
import pandas as pd

# Calculate the correlation between CO2 dissolved and Quantity in litres
correlation_coefficient = df['CO2 dissolved'].corr(df['Quantity (lts.)'])

# Round off to 2 decimal places
correlation_coefficient_rounded = round(correlation_coefficient, 2)

# Display the result
print(f"The correlation between CO2 dissolved and Quantity in litres is: {correlation_coefficient_rounded}")
```

```python
# Calculate Q1 and Q3 for CO2 dissolved
Q1_CO2 = df['CO2 dissolved'].quantile(0.25)
Q3_CO2 = df['CO2 dissolved'].quantile(0.75)

# Calculate IQR for CO2 dissolved
IQR_CO2 = Q3_CO2 - Q1_CO2

# Calculate lower and upper cutoffs for CO2 dissolved
lower_cutoff_CO2 = round(Q1_CO2 - 1.5 * IQR_CO2, 2)
upper_cutoff_CO2 = round(Q3_CO2 + 1.5 * IQR_CO2, 2)

# Display the results
print(f"Lower Cutoff for CO2 dissolved: {lower_cutoff_CO2}")
print(f"Upper Cutoff for CO2 dissolved: {upper_cutoff_CO2}")
```

```python
# Exclude outliers based on the identified cutoffs for CO2 dissolved
df_no_outliers_CO2 = df[(df['CO2 dissolved'] >= lower_cutoff_CO2) & (df['CO2 dissolved'] <= upper_cutoff_CO2)]

# Calculate the mean value for the "CO2 dissolved" variable after treating outliers
mean_CO2_no_outliers = df_no_outliers_CO2['CO2 dissolved'].mean()

# Display the mean value
print(f"The mean value for CO2 dissolved variable after treating outliers is: {mean_CO2_no_outliers:.2f}")
```

```python
import math

# Given probabilities for Assembly line B
p_line_B = 0.21

# Set the values
n_trials = 50
k_successes = 10

# Calculate the binomial probability
binomial_probability_line_B = (
    (math.factorial(n_trials) / (math.factorial(k_successes) * math.factorial(n_trials - k_successes)))
    * (p_line_B ** k_successes)
    * ((1 - p_line_B) ** (n_trials - k_successes))
)

# Round off to 2 decimal places
binomial_probability_line_B_rounded = round(binomial_probability_line_B, 2)

# Display the result
print(f"The probability that 10 bottles crossed the time limit out of 50 on Assembly line B is: {binomial_probability_line_B_rounded}")
```

```python
import math

# Given probability for Assembly line B
p_line_B = 0.21

# Set the values
n_trials = 50
k_successes_min = 10  # Minimum number of successes for "at least 10"

# Calculate the cumulative binomial probability for at least 10 successes
binomial_probabilities_line_B = sum(
    (math.factorial(n_trials) / (math.factorial(k) * math.factorial(n_trials - k)))
    * (p_line_B ** k)
    * ((1 - p_line_B) ** (n_trials - k))
    for k in range(k_successes_min, n_trials + 1)
)

# Round off to 2 decimal places
probability_at_least_10_line_B = round(binomial_probabilities_line_B, 2)

# Display the result
print(f"The probability of at least 10 bottles crossing the time limit out of 50 on Assembly line B is: {probability_at_least_10_line_B}")
```

```python
from scipy.stats import norm

# Calculate the mean and standard deviation for the Quantity variable after removing outliers
mean_quantity_no_outliers = df['Quantity (lts.)'].mean()
std_quantity_no_outliers = df['Quantity (lts.)'].std()

# Set the values for penalization and spillage limits
penalization_limit = 1.90
spillage_limit = 2.10

# Calculate the z-scores for penalization and spillage limits
z_score_penalization = (penalization_limit - mean_quantity_no_outliers) / std_quantity_no_outliers
z_score_spillage = (spillage_limit - mean_quantity_no_outliers) / std_quantity_no_outliers

# Calculate the probability for penalization and spillage using the normal distribution CDF
probability_penalization = norm.cdf(z_score_penalization)
probability_spillage = 1 - norm.cdf(z_score_spillage)

# Calculate the combined probability
probability_combined = probability_penalization + probability_spillage

# Round off to 2 decimal places
probability_combined_rounded = round(probability_combined, 2)

# Display the result
print(f"The probability of the bottle content being either penalized or causing spillage is: {probability_combined_rounded}")
```

```python
from scipy.stats import norm

# Given values
mean_bill = 850  # Mean monthly cell phone bill
std_dev_bill = 150  # Standard deviation of the monthly cell phone bill

# Amounts for which we want to calculate the probabilities
amount_more_than_1200 = 1200
lower_limit = 750
upper_limit = 1200

# Calculate the z-scores
z_score_more_than_1200 = (amount_more_than_1200 - mean_bill) / std_dev_bill
z_score_lower = (lower_limit - mean_bill) / std_dev_bill
z_score_upper = (upper_limit - mean_bill) / std_dev_bill

# Calculate the probabilities using the normal distribution CDF
probability_more_than_1200 = 1 - norm.cdf(z_score_more_than_1200)
probability_between_750_and_1200 = norm.cdf(z_score_upper) - norm.cdf(z_score_lower)

# Round off to 2 decimal places
probability_more_than_1200_rounded = round(probability_more_than_1200, 2)
probability_between_750_and_1200_rounded = round(probability_between_750_and_1200, 2)

# Display the results
print(f"The probability that a randomly selected cell phone bill is more than Rs 1200 is: {probability_more_than_1200_rounded}")
print(f"The probability that a randomly selected cell phone bill is between Rs 750 and Rs 1200 is: {probability_between_750_and_1200_rounded}")
```
